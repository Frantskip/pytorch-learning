```python
import torch
import numpy as np
```

# 张量创建
## 1 创建张量

```python
tensor1 = torch.tensor(10.0)
print(tensor1)
print(tensor1.dtype)
print(tensor1.size())
```

```python
tensor2 = torch.tensor(np.array([[1, 2, 3],[4, 5, 6]]))
print(tensor2)
print(tensor2.dtype)
print(tensor2.size())
```

```python
# 2 创建指定类型的张量
tensor1 = torch.Tensor(3,2,4)
print(tensor1)
print(tensor1.dtype)
print(tensor1.size())
```

```python
tensor2 = torch.Tensor(np.array([[1, 2, 3, 4],[4, 5, 6, 7]]))
print(tensor2)
print(tensor2.dtype)
```

```python
# int(int32) 类型
tensor1 = torch.IntTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.int32)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# int(int64) 类型
tensor1 = torch.LongTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.int64)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# short (int16) 类型
tensor1 = torch.ShortTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.int16)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# int(int8) 类型
tensor1 = torch.ByteTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.uint8)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# float(float32) 类型
tensor1 = torch.FloatTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.float32)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# float(float64) 类型
tensor1 = torch.DoubleTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.float64)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# half(float16) 类型
tensor1 = torch.HalfTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.float16)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# bool 类型
tensor1 = torch.BoolTensor(2,3)
tensor2 = torch.tensor([1,2,3],dtype=torch.bool)
print(tensor1.dtype)
print(tensor2.dtype)
print(tensor1)
print(tensor2)
```

```python
# 指定区间的张量创建
# 1 torch.arange() 函数
# 功能：返回一个1维张量，包含从start到end（不包含end）的以step为步长的序列
# 语法：torch.arange(start, end, step)
# 参数：
# start：序列的起始值，默认值为0
# end：序列的结束值（不包含在序列中）
# step：序列中每个元素之间的步长，默认值为1
# 返回值：1维张量
tensor1 = torch.arange(10,30,2)
print(tensor1)
tensor2 = torch.arange(6)
print(tensor2)
```

```python
# 2 torch.linspace() 函数
# 功能：返回一个1维张量，包含从start到end（包含end）的num个等间隔的值
# 语法：torch.linspace(start, end, num)
# 参数：
# start：序列的起始值
# end：序列的结束值（包含在序列中）
# steps：序列中元素的数量，默认值为100
# 返回值：1维张量
tensor1 = torch.linspace(10,30,6)
print(tensor1)
```

```python
# 3 torch.logspace() 函数
# 功能：返回一个1维张量，包含从start到end（包含end）的num个等间隔的值，这些值是10的幂次

# 参数：
# start：序列的起始值，默认值为0
# end：序列的结束值（包含在序列中）
# steps：序列中元素的数量，默认值为100
# 返回值：1维张量
tensor1 = torch.logspace(1,3,3,base=2)
print(tensor1)
```

```python
# 按数值
tensor1 = torch.zeros(2,3)
print(tensor1)
print(tensor1.dtype)
```

```python
tensor2 = torch.ones_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.full((2,3),6.5)
print(tensor1)
print(tensor1.dtype)
```

```python
tensor2 = torch.empty_like(tensor1)
print(tensor2)
print(tensor2.dtype)
```

```python
tensor1 = torch.eye(3,5)
print(tensor1)
print(tensor1.dtype)
```

```python
# 随机张量创建
tensor1 = torch.rand(2,3)
print(tensor1)
print(tensor1.dtype)
tensor2 = torch.rand_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.randint(low=0, high=100, size=(2,3))
print(tensor1)
tensor2 = torch.randint_like(tensor1, low=0, high=100)
print(tensor2)
```

```python
# randn 随机生成标准正态分布（均值为0，标准差为1）的张量
tensor1 = torch.randn(4,3)
print(tensor1)
tensor2 = torch.randn_like(tensor1)
print(tensor2)
```

```python
# torch.normal 随机生成正态分布（均值为mean，标准差为std）的张量
tensor1 = torch.normal(5,2 ,size=(2,3))
print(tensor1)
tensor2 = torch.normal(mean=5, std=2, size=tensor1.shape)
print(tensor2)
```

```python
# torch.randperm 随机生成一个从0到n-1的随机排列的张量
tensor1 = torch.randperm(10)
print(tensor1)
```
